```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("population.csv")
```

```python
data
```

```python
data.info()
```

```python
#List of contries for the graph.
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2023'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2023')
plt.title('Population of selected countries in 2023')
plt.xticks(rotation=45, ha='right')
plt.show()
```
